```python
from decimal import *
import matplotlib.pyplot as plt, time

class funcion:
    def iniciar(ninicio: Decimal, nfin:Decimal) -> 0:
        global listavalores, listavalores_x, listaintervalo, solucion, tiempoinicio
        listaintervalo = []
        listavalores = []
        listavalores_x = []
        solucion = True
        getcontext().prec = 7
        if ninicio == nfin == 0:
            while True:
                try:
                    ninicio = Decimal(input("Inicio del intervalo de la funcion: ").replace(',','.'))
                    ninicio * 1
                    nfin = Decimal(input("Fin del intervalo de la funcion: ").replace(',','.'))
                    nfin * 1
                    if nfin != ninicio:
                        if nfin > ninicio:
                            listaintervalo.append(ninicio)
                            listaintervalo.append(nfin)
                            tiempoinicio = time.time()
                            break
                        print('El número de inicio no puede ser más grande que el de final ')
                    print("El número de inicio no piede ser igual que el de final")
                except:
                    pass

    def condicion(inicio, fin):
        global solucion
        if fin-inicio <= 0.2:
            solucion = False
    
    def salir(exito:int):
        tiempofin = time.time()
        tiempototal = tiempofin - tiempoinicio
        print(tiempototal)
        print(count)
        if exito == 0:
            print("La función no tiene punto de corte")
        quit()
        exit()

    def contador(start:Decimal, end:Decimal, suma:Decimal) -> 0:
        global listanumero
        #start = round(Decimal.normalize(start), 3)
        #end = round(Decimal.normalize(end), 3)
        #print(start)
        #print(end)
        listanumero = []
        control = True
        cuenta = start
        while control:
            cuenta += suma
            listanumero.append(cuenta)
            if cuenta >= end:
                control = False
    
    def calcular_polinomio(x):
        resultado = Decimal('0')
        # INGRESE EL POLINOMIO
        polinomio = [x**3, -7*x, 6]
        # INGRESE EL POLINOMIO 
        for termino in polinomio:
            resultado += termino
        if -10**-6 <= resultado <= 10**-6:
            print(f"En el valor de x: {x} la función tiene de valor: {Decimal.normalize(resultado)}")
            funcion.salir(1)
        print(x)
        listavalores.append(resultado)
        listavalores_x.append(x)
        return resultado
    
    def calcular_intervalo():
        global count
        iniciox = listavalores_x[0]
        finx =  listavalores_x[1]
        iniciovalor = listavalores[0]
        finvalor = listavalores[1]
        count = 0
        while solucion:
            mediox = Decimal((finx+iniciox)/2)
            mediovalor = funcion.calcular_polinomio(mediox)
            valor = iniciovalor*mediovalor
            valor2 = mediovalor*finvalor
            if valor < 0:
                finx = mediox
                finvalor = mediovalor
            elif valor2 <= 0:
                iniciox = mediox
                iniciovalor = mediovalor
            else:
                funcion.contador(iniciox, finx, Decimal('0.5'))
                for numero in listanumero:
                    funcion.calcular_polinomio(numero)
                print("No tiene punto de corte")
                funcion.salir()
            count += 1
            funcion.condicion(iniciox, finx)
        funcion.contador(iniciox, finx, Decimal('0.00001'))
        for numero in listanumero:
              funcion.calcular_polinomio(numero)
        funcion.salir(0)

    def funcion():
        funcion.iniciar(0, 0)
        for numero in listaintervalo:
            funcion.calcular_polinomio(numero)
        funcion.calcular_intervalo()
if __name__ == "__main__":
    funcion.funcion()
```
